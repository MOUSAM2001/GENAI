## Read pdf file

```python
from google import genai
from google.genai import types
import httpx

client = genai.Client()

# document URL
doc_url = "https://publications.lib.chalmers.se/records/fulltext/238913/238913.pdf"

doc_data = httpx.get(doc_url).content

prompt=" what are the key aspects of demand planning mentioned in this document?"
response= client.models.generate_content(
    model= "gemini-2.5-flash",
    contents= [
        types.Part.from_bytes(
            data=doc_data,
            mime_type="application/pdf"
        ), prompt
    ])
print("Response: ", response.text)
```

## Read PDF from Storage

```python
from google import genai
from google.genai import types
import pathlib

client = genai.Client()

filepath= pathlib.Path("/workspaces/GENAI/campusx/files/Mousam_Kumar_2025-11-30.pdf")

prompt="what this person is doing"

response= client.models.generate_content(
    model= "gemini-2.5-flash",
    contents= [
        types.Part.from_bytes(
            data=filepath.read_bytes(),
            mime_type="application/pdf"
        ), prompt
    ])

print(response.text)
```
